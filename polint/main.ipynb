## Question 2



```python
def get_position(x, xx):
    l = 0
    r = len(xx) - 1
    while l<=r:
        mid = (l + r)//2
        if x < xx[mid]:
            r =  mid - 1
        elif x > xx[mid]:
            l =  mid + 1
        else:
            return mid
    return l

def nearest_n(x, xx, n):
    if len(xx) == n:
        return xx
    
    pos = get_position(x, xx)
    if n%2 != 0:
        if pos < n//2 + 1:
            return xx[:n]
        elif pos > len(xx) - n//2 - 1:
            return xx[-n:]
        else:
            le = x - xx[pos - n//2 - 1]
            re = xx[pos + n//2] - x

            if le <= re:
                return xx[pos - n//2 - 1: pos + n//2]
            else:
                return xx[pos - n//2: pos + n//2 + 1]
    
    if n%2 == 0:
        if pos < n//2:
            return xx[:n]
        elif pos > len(xx) - n//2 - 1:
            return xx[-n:]
        else:
            return xx[pos - n//2: pos + n//2]
```

```python
# sampling 30 points from sin(x + x^2) from 0 to 1
xx = np.linspace(0., 1., 30)
yy = f1(xx)

# getting the points to interpolate
x = np.linspace(-0.5, 1.5, 200)
```

We see that the error in ther interpolated region has reduced. We see this from $f(x) - p_n(x) = \dfrac{f^{(n+1)}(\xi)\Pi_{i=0}^{n}(x-x_i)}{(n+1)!}$. Since the order $n$ is unchanged $n=4$. The difference comes from the term $\Pi_{i=0}^{n}(x-x_i)$ since $x_i$ in the second case are closer to the $x$ than in the first case, the error in the second case is lower.

The total error in interpolation is the sum of mathematical error and the arithmetic error due to the floating point approximation $TE_n(x) = ME_n(x) + AE_n(x)$. $TE_n(x) = f(x) - \tilde{p_n(x)}$ and $ME_n(x) = f(x) - p_n(x)$ and $AE_n(x) = p_n(x) - \tilde{p_n(x)}$, where $p_n(x)$ is the interpolating polynomial for $\{x_i, f(x_i)\}_{i=0}^{n}$ and $\tilde{p_n(x)}$ is the interpolating polynomial for $\{x_i, FL(f(x_i))\}_{i=0}^{n}$, where $FL(f(x_i))$ is the floating point approximation.

Now we derive an upper bound for the Arithmetic error
$AE_n(x) = p_n(x) - \tilde{p_n(x)} = \sum_{k=0}^{n}(f(x_k) - FL(f(x_k)))L_k$, where $\{L_k\}_{k}$ are the Lagrange polynomials. Now define $\epsilon_i := f(x_i) - FL(f(x_i))$ and $ ||\epsilon||_{\infty} = \max{|\epsilon_i| : i=0,1,\ldots,n}$. Now $\forall x \in \mathcal{I}$ where $\mathcal{I}$ is the interpolating region. 
$|AE_n(x)| = |p_n(x) - \tilde{p_n(x)}| \leq ||\epsilon||_{\infty} \sum_{i=0}^{n}||L_i||_{\infty,\mathcal{I}}$.

It can be shown that $\sum_{i=0}^{n}||L_i||_{\infty,\mathcal{I}}$ when the table points are equally spaced, can grow exponentially with $n$. $\sum_{i=0}^{n}||L_i||_{\infty,\mathcal{I}} \leq 2^{n}$.

Therefore the Arithmetic error has the potential to grow large with $n$ theoretically. And in practice it almost always grow exponentially with $n$.

```python
interpolation_orders = [0, 1, 2, 5, 9, 10, 15, 20, 25, 29]

for k in interpolation_orders:
    # doing kth order interpolation with nearest k+1 points
    y, dy = [], []

    for w in x:
        xx_nearest = nearest_n(w, xx, k+1)
        yy_nearest = f1(xx_nearest)

        y_dy = p.polint(xx_nearest, yy_nearest, w)
        y.append(y_dy[0])
        dy.append(y_dy[1])
    
    # plotting the interpolated values
    plt.figure(1, figsize=(10, 6))
    plt.plot(xx, yy, 'ro', x, y, 'r', x, f1(x), 'g')
    plt.title(f"{k}th order interpolation of $sin(x + x^2)$ - 30 points")
    plt.legend(["Table values", "Interpolated values", "True function"])
    plt.grid(True)

    # plotting the semilog plot of the errors
    plt.figure(2, figsize=(10, 6))
    plt.semilogy(x, np.abs(dy), 'r', x, np.abs(f1(x)-y), 'g')
    plt.title("Error plot")
    plt.legend(["Estimated Error","True error"])
    plt.grid(True)

    plt.show()
```

We can bound the mathematical error for this particular function $f(x) = \sin(x^2+x)$ using the expression for the mathematical error. $\Pi_{i=0}^{n}(x-x_i)$ can be bounded in the interpolation region as $\Pi_{i=0}^{n}(x-x_i) \leq \Pi_{i=0}^{n}|x_0-x_n| = \Pi_{i=0}^{n}1 = 1$ and we observe that the factorial term in the denominator eventually overtakes the numerator term leading to a decaying behaviour(see the plot below), hence with veiwing only the mathematical error we expect the error in the interpolation region to decrease with $n$ as the function is analytical. But this is not we observe, we see that after certain point the error in the interpolating region starts to rise. 

And in the extrapolating region, the term $\Pi_{i=0}^{n}(x-x_i)$ is not upper-bounded and can grow to very large number as $x$ gets further and further from the nodal points. And also the $\textbf{Runge Phenomenon}$ near the edges of the table values affect the extrapolating error as the table points are equally spaced(though this can be reduced using $\texttt{chebyshev points}$)

```python
x = sp.symbols('x')
f = sp.sin(x + x**2)

grid = np.linspace(0, 1, 1000)

max_values = []

for n in range(3, 22):
    nth_derivative = sp.diff(f, x, n)
    nth_derivative_func = sp.lambdify(x, nth_derivative, 'numpy')
    max_value = np.max(np.abs(nth_derivative_func(grid)))
    max_values.append(max_value)
    
max_values = np.array(max_values)
fac_values = np.array([math.factorial(n+1) for n in range(3,22)])

values = max_values/fac_values

plt.semilogy(range(3, 22), values, marker='o')
plt.xlabel('n (Order of Derivative)')
plt.ylabel('Maximum Value of nth Derivative')
plt.title('Maximum of nth Derivative of sin(x + x^2) over [0,1]')
plt.grid(True)
plt.show()
```

```python
interpolation_orders = range(3,21)
x = np.linspace(-0.5, 1.5, 200)

intra_max_est, intra_max_true, extra_max_est, extra_max_true = [], [], [], []
for k in interpolation_orders:
    # doing kth order interpolation with nearest (k+1) points
    int_dy_est_max, int_dy_true_max = -np.inf, -np.inf
    ext_dy_est_max, ext_dy_true_max = -np.inf, -np.inf
    for w in x:
        xx_nearest = nearest_n(w, xx, k+1)
        yy_nearest = f1(xx_nearest)

        y, dy_est = p.polint(xx_nearest, yy_nearest, w)
        dy_est = np.abs(dy_est)
        dy_true = np.abs(f1(w) - y)
        if w < 0. or w > 1.:
            if dy_est > ext_dy_est_max:
                ext_dy_est_max = dy_est
            if dy_true > ext_dy_true_max:
                ext_dy_true_max = dy_true
        else:
            if dy_est > int_dy_est_max:
                int_dy_est_max = dy_est
            if dy_true > int_dy_true_max:
                int_dy_true_max = dy_true
    intra_max_est.append(int_dy_est_max)
    intra_max_true.append(int_dy_true_max)
    extra_max_est.append(ext_dy_est_max)
    extra_max_true.append(ext_dy_true_max)
```

```python
# plotting the max error vs the interpolation order in the interpolation region
plt.figure(1, figsize=(10, 6))
plt.semilogy(interpolation_orders, intra_max_est, 'r', interpolation_orders, intra_max_true, 'g')
plt.title("Max-error vs Interpolation Order - Interpolated Region")
plt.legend(['Estimated','True'])
plt.xticks(interpolation_orders)
plt.xlabel('Interpolation Order')
plt.ylabel('Log of max error')
plt.grid(True)

# plotting the max error vs the interpolation order in the extrapolation region
plt.figure(2, figsize=(10, 6))
plt.semilogy(interpolation_orders, extra_max_est, 'r', interpolation_orders, extra_max_true, 'g')
plt.title("Max-error vs Interpolation Order - Extrapolated Region")
plt.legend(['Estimated','True'])
plt.xticks(interpolation_orders)
plt.xlabel('Interpolation Order')
plt.ylabel('Log of max error')
plt.grid(True)

plt.show()
```

## Question 1



We know that

$C_{m,i} = P_{i,\ldots,(i+m)} - P_{i,\ldots,(i+m-1)}$ 

$D_{m,i} = P_{i,\ldots,(i+m)} - P_{(i+1),\ldots,(i+m)}$ 

Now, $D_{m+1,i} = P_{i,\ldots,(i+m+1)} - P_{(i+1),\ldots,(i+m+1)} = \dfrac{(x - x_{i+m+1})P_{i,\ldots,(i+m)} + (x_i - x)P_{(i+1),\ldots,(i+m+1)}}{x_i - x_{i+m+1}} - P_{(i+1),\ldots,(i+m+1)}$

= $\dfrac{(x_{i+m+1} - x)(P_{(i+1),\ldots,(i+m+1)} - P_{i,\ldots,(i+m)})}{x_i - x_{i+m+1}}$

= $\dfrac{(x_{i+m+1} - x)(P_{(i+1),\ldots,(i+m+1)} - P_{(i+1),\ldots,(i+m)} + P_{(i+1),\ldots,(i+m)} - P_{i,\ldots,(i+m)})}{x_i - x_{i+m+1}}$

= $\dfrac{(x_{i+m+1} - x)(C_{m,i+1} - D_{m,i})}{x_i - x_{i+m+1}}$

similarly 

$C_{m+1,i} = P_{i,\ldots,(i+m+1)} - P_{i,\ldots,(i+m)} = \dfrac{(x - x_{i+m+1})P_{i,\ldots,(i+m)} + (x_i - x)P_{(i+1),\ldots,(i+m+1)}}{x_i - x_{i+m+1}} - P_{i,\ldots,(i+m)}$

= $\dfrac{(x_{i} - x)(P_{(i+1),\ldots,(i+m+1)} - P_{i,\ldots,(i+m)})}{x_i - x_{i+m+1}}$

= $\dfrac{(x_{i} - x)(P_{(i+1),\ldots,(i+m+1)} - P_{(i+1),\ldots,(i+m)} + P_{(i+1),\ldots,(i+m)} - P_{i,\ldots,(i+m)})}{x_i - x_{i+m+1}}$

= $\dfrac{(x_{i} - x)(C_{m,i+1} - D_{m,i})}{x_i - x_{i+m+1}}$

```python
def neville(xx, yy, x):
    n = len(xx)
    Q = np.zeros((n, n))
    for i in range(n):
        Q[i][0] = yy[i]
    for i in range(1, n):
        for j in range(1, i + 1):
            Q[i][j] = ((x - xx[i - j]) * Q[i][j - 1] 
                       - (x - xx[i]) * Q[i - 1][j - 1]) / (xx[i] - xx[i - j])
    return Q[n-1][n-1], Q

def plot_tableu(Q):
    n = Q.shape[0]
    G = nx.DiGraph()
    pos = {}
    labels = {}
    for i in range(n):
        for j in range(i + 1):
            node = (i, j)
            G.add_node(node)
            pos[node] = (j, i)
            labels[node] = f"{Q[i][j]:.6f}"
            if j > 0:
                parent1 = (i, j - 1)
                parent2 = (i - 1, j - 1)
                G.add_edge(parent1, node)
                G.add_edge(parent2, node)
    plt.figure(figsize=(10, 8))
    nx.draw(G, pos, labels=labels, with_labels=True, node_size=3000, node_color="lightblue", font_size=10)
    plt.gca().invert_yaxis()
    plt.show()
```

```python
# sampling 5 points from sin(x + x^2) from 0 to 1
xx = np.linspace(0., 1., 5)
yy = f1(xx)

x = 0.33

y, Q = neville(xx, yy, x)
print(f"The approximated value of f({x}) is: {y:.9f}")
```

```python
k = 10
n_order = range(1,100)

norunge_err, runge_err = [], []

for n in n_order:
    xx = np.linspace(0,10,n+1)
    yy = np.sin(k*xx)
    
    x_no_runge = np.linspace(3,7,1000)
    x_runge =  np.linspace(0,10,1000)
    
    y_runge = np.array([p.polint(xx, yy, w) for w in x_runge])[:,0]
    y_no_runge = np.array([p.polint(xx, yy, w) for w in x_no_runge])[:,0]
    dy_true_runge = np.abs(np.sin(k*x_runge) - y_runge)
    dy_true_norunge = np.abs(np.sin(k*x_no_runge) - y_no_runge)
    norunge_err.append(np.max(np.abs(dy_true_norunge)))
    runge_err.append(np.max(dy_true_runge))

plt.figure(figsize=(10,8))
plt.semilogy(n_order, norunge_err, 'r', n_order, runge_err, 'g')
plt.legend(['Without Runge Phenomenon', 'With Runge Phenomenon'])
plt.grid(True)
plt.show()
```

```python
k_vals = [2, 5, 10, 20, 30]
n_order = range(1, 100)

plt.figure(figsize=(10, 8))

for k in k_vals:
    norunge_err = []

    for n in n_order:
        xx = np.linspace(0, 10, n + 1)
        yy = np.sin(k * xx)

        x_no_runge = np.linspace(3, 7, 1000)

        y_no_runge = np.array([p.polint(xx, yy, w) for w in x_no_runge])[:, 0]
        dy_true_norunge = np.abs(np.sin(k * x_no_runge) - y_no_runge)
        norunge_err.append(np.max(np.abs(dy_true_norunge)))

    plt.semilogy(n_order, norunge_err, label=f'k = {k}')

plt.legend(title='k values')
plt.xlabel('n')
plt.ylabel('Maximum Interpolation Error')
plt.grid(True)
plt.title('No Runge - Interpolation Error vs. n for Different k Values')
plt.show()
```
