### Question - 2

We will graph the integrand of this function to understand its characteristics

```python
v, k, g = 3, 2.7, 1.2
```

```python
def integrand(u):
    if type(u).__name__ == 'ndarray':
        y = np.array([integrand(i) for i in u])
        return y
        
    if u<1.0:
        return 2*jv(v, k*u)**2*u
        
    j = 2*abs(jv(v, k)/kv(v, g))**2
    return j*kv(v, u*g)**2*u
```

```python
true_val = (jv(v, k)**2 - jv(v+1, k)*jv(v-1, k)) + abs(jv(v, k)/kv(v, g))**2*(kv(v+1, g)*kv(v-1, g) - kv(v, g)**2) 
```

```python
print(f'The true value of the integrand is : {true_val}')
```

### Question - 3

We will now analytically bound the upperlimit $\infty$ with some $u_{max}$ such that we could compute the integral numerically

![q3](q3.jpeg)

We can test whether the above is a good approximation by looking at the error for large $u$.

```python
second_part_val = abs(jv(v, k)/kv(v, g))**2*(kv(v+1, g)*kv(v-1, g) - kv(v, g)**2)
```

```python
print(f'The integral value for the interval : (1, inf) : {second_part_val}')
```

```python
def approx_func(x):
    c = (np.pi/g)*(abs(jv(v, k)/kv(v, g))**2)
    return c*np.exp(-2*g*x)
```

```python
a1 = np.linspace(1, 20, 1000)
y1 = integrand(a1)
```

```python
plt.plot(a1, y1, label='True Function')
plt.plot(a1, approx_func(a1), label='Approximated Function')
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.xlabel('u')
plt.ylabel('y')
plt.title("Approximated Function vs True Function")
plt.legend()
plt.show()
```

We can see that the approximated function does a **bad** job at capturing the function's characteristics near $u=1$, but we are not interested in the function's behavior near but rather at large values of **$u$**.

We can measure how well the function behaves for large values of $u$ by looking at the error in a log-plot

```python
a1 = np.logspace(0, 4, endpoint=True)
y1 = integrand(a1)

plt.loglog(a1, abs(y1-approx_func(a1)), label='Error')
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.xlabel('u')
plt.ylabel('y')
plt.title("Error vs $u$")
plt.legend()
plt.show()
```

The above log-log plot shows that the decrease in error is **exponential** proving that the approximation works very well

We will now find the estimated value of $u_{max}$

```python
c = (np.pi/g)*(abs(jv(v, k)/kv(v, g))**2)
umax = -np.log(pow(10, -15)*2*g/c)*(1/2*g)

print(f'The estimated value of u_max : {umax}')
```

### Question - 5

We will now implement the **trapzd** function from the *romberg* module

```python
def trapzd(func, a, b, n):
    func_calls = 2 if n==1 else 2 + pow(2, n-1) - 1
    s_in = 0
    for i in range(1, n+1):
        s_in = romberg.trapzd(func, a, b, s_in, i)

    # global count;
    # func_calls = count
    # count = 0
    return s_in, func_calls
```

```python
trapzd(integrand, 0, 1000, 18)
```

```python
trapzd_errors = []
trapzd_calls = []

for i in range(2, 20):
    y, func_calls = trapzd(integrand, 0, umax, i)
    trapzd_errors += [abs(y-true_val)]
    trapzd_calls += [func_calls]

plt.loglog(trapzd_calls, trapzd_errors, marker='o', markerfacecolor='white')
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.xlabel("No. of function calls")
plt.ylabel("Error")
plt.title("Error vs No. of function calls - (trapzd)")
plt.show()
```

```python
trapzd_split_errors = []
trapzd_split_calls = []

for i in range(2, 20):
    y1, func_calls1 = trapzd(integrand, 0, 1, i)
    y2, func_calls2 = trapzd(integrand, 1, umax, i)
    trapzd_split_errors += [abs(y1+y2-true_val)]
    trapzd_split_calls += [func_calls1+func_calls2]

plt.loglog(trapzd_split_calls, trapzd_split_errors, marker='o', markerfacecolor='white')
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.xlabel("No. of function calls")
plt.ylabel("Error")
plt.title("Error vs No. of function calls - (trapzd : $(0,1), (1,\infty)$)")
plt.show()
```

### Question - 12

We will now rewrite the python interface so as to change **h -> h/2** to **h -> h/3** after each iteration. 

First we need to implement a *new_trapzd* algorithm that implements divide by **3** method.

```python
def new_trapzd(func, a, b, s_old, n, div=3):
    """
    Parameters:
    func : Evaluation Function (func->float)
    a : Lower limit (int, float)
    b : Upper Limit (int, float)
    s_old : Previously computed Integral value (float)
    n : Current iteration (int)
    -------------------------------------------------
    Returns:
    s : Computed Integral Value for the current iteration(float)
    calls : No. of function calls (int)
    """
    h = (b-a)

    if n==1: return 0.5*(func(a)+func(b))*h

    s_new = 0
    k = pow(div, n)
    for cnt in range(1, k):
        if cnt%div == 0:
            continue
        s_new += func(a + cnt*h/k)

    return s_new*h/k + s_old/div    
    


def new_trapzd_wrapper(func, a, b, n, div=3):
    """
    A wrapper function for easy usage of the divide-by-3 trapzd defined
    """
    s = 0
    func_calls = pow(div, n-1) + 1
    
    for i in range(1, n):
        s = new_trapzd(func, a, b, s, i, div)

    return s, func_calls


# def qromb_3(func, a, b, eps=pow(10, -6), order=5):
#     h = (b-a)
#     imax = 20
#     xx = []
#     yy = []
#     s = 0

#     for i in range(imax)
#     while error > eps or i < order:
#         s = new_trapzd(func, a, b, s, i+1) # i+1 since i goes from '0' to 'order-1'
#         # print(s)
#         if i==0:
#             error = float('inf')
#         else:
#             error = abs(s-temp_s)
#         xx += [(h/pow(3, i)**2)]
#         yy += [s]
#         temp_s = s
#         i += 1
        

#     y, err = romberg.polint(xx[-order:], yy[-order:], 0)
#     func_calls = 1 + pow(3, i)

#     return y, err, func_calls 

def qromb_3(func, a, b, eps=pow(10, -6), k=5, div=3):
    h = (b-a)
    imax = 20
    xx = []
    yy = []
    s = 0

    for i in range(imax):
        j = i+1
        s = new_trapzd(func, a, b, s, j, div) # i+1 since i goes from '0' to 'order-1'
        xx += [(h/pow(div, i))**2]
        yy += [s]
        if j > k:
            y, err = romberg.polint(xx[-(k):], yy[-(k):], 0)
            if abs(err) <= eps*abs(y):
                # print(len(xx[-(k+1):-1]))
                numcalls = pow(div, i) + 1
                return y, err, numcalls

    numcalls = -1
    return y, err, numcalls
```
