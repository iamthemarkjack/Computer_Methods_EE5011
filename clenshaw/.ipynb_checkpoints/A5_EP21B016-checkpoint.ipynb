### Question 1 : A *C* program to compute the root using the exact formula

```python
import ctypes
exact_file = r"./quadratic_exact.so"
float_file = r"./quadratic_float.so"
```

```python
quadratic_exact = ctypes.CDLL(exact_file)
quadratic_float = ctypes.CDLL(float_file)
```

```python
class exactRoots(ctypes.Structure):
    _fields_ = [
        ('root1_r', ctypes.c_double),
        ('root1_i', ctypes.c_double),
        ('root2_r', ctypes.c_double),
        ('root2_i', ctypes.c_double)
    ]

quadratic_exact.find_roots.argtypes = [ctypes.c_double]
quadratic_exact.find_roots.restype = exactRoots

def find_exact_roots(alpha):
    c_in = ctypes.c_double(alpha)
    res = quadratic_exact.find_roots(c_in)
    return res.root1_r + 1j*res.root1_i, res.root2_r + 1j*res.root2_i
```

```python
class floatRoots(ctypes.Structure):
    _fields_ = [
        ('root1_r', ctypes.c_float),
        ('root1_i', ctypes.c_float),
        ('root2_r', ctypes.c_float),
        ('root2_i', ctypes.c_float)
    ]

quadratic_float.find_float_roots.argtypes = [ctypes.c_float]
quadratic_float.find_float_roots.restype = floatRoots

quadratic_float.find_acc_roots.argtypes = [ctypes.c_float]
quadratic_float.find_acc_roots.restype = floatRoots

def find_float_roots(alpha):
    c_in = ctypes.c_float(alpha)
    res = quadratic_float.find_float_roots(c_in)
    return res.root1_r + 1j*res.root1_i, res.root2_r + 1j*res.root2_i

def find_acc_roots(alpha):
    c_in = ctypes.c_float(alpha)
    res = quadratic_float.find_acc_roots(c_in)
    return res.root1_r + 1j*res.root1_i, res.root2_r +  1j*res.root2_i
```

### Question 2&3 : Computing the errors (assuming part 1 to be exact) of roots found with float precision and exact formula and with the more accurate formula

```python
x = np.logspace(-5, 5, 1000)
y = np.array([find_exact_roots(i) for i in x])
y1 = np.array([find_float_roots(i) for i in x])
y2 = np.array([find_acc_roots(i) for i in x])
```

# Stable & Unstable Series

```python
from scipy.special import jv
```

```python
def exact_summation(n, x):
    l = np.cumsum(np.array([1/(i+1)*jv(i, x) for i in range(n+1)]))
    return l, l[-1]

def forward_series(n, x):
    a, b = jv(0, x), jv(1, x)
    if n==0:
        return a
    if n==1:
        return b

    for i in range(1, n):
        c = 2*i*b/x -a
        a = b
        b = c
        
    return c    


def forward_summation(n, x):
    l = np.cumsum(np.array([1/(i+1)*forward_series(i, x) for i in range(n+1)]))
    return l, l[-1]
```

```python
n = 10
x = 1.5

print(f'Exact Bessel : {jv(n, x)}')
print(f'Forward Bessel : {forward_series(n, x)}')
```

```python
n = 40
x = 1.5

print(f'Exact Summation : {exact_summation(n, x)[1]}')
print(f'Summation using forward series : {forward_summation(n, x)[1]}')
```

### Implementing the backward series

```python
def backward_series(n, x):
    j_series = np.zeros(n+2)
    j_series[1] = 1
    for i in range(2, n+2):
        j_series[i] = 2*(n+2-i)*j_series[i-1]/x - j_series[i-2]
    
    alpha = j_series[-1]/jv(0, x)
    return j_series[::-1][:-1]/alpha
```

```python
def backward_summation(n, x):
    series = backward_series(n, x)
    den = np.array(range(1, n+2))
    series /= den
    res = np.cumsum(series)
    return res, res[-1]
```

```python
n = 40
x = 1.5

l1, res1 = exact_summation(60, x)
l2, res2 = backward_summation(60, x)
err_1 = abs(l1[:41] - l2[:41])
plt.semilogy(range(n+1), err_1)
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.title(f'Error ($x={x}$)- (Backward Recursion)')
plt.ylabel('Error')
plt.xlabel('n')
plt.show()
```

# Clenshaw algorithm

### Overview of *Clenshaw Algorithm*

Say
$$ S(x) = \sum_{n=0}^{N} a_n f_n(x) $$
where, $f_n(x)$ satisfies the recurrence relation:
$$ f_{n+1}(x) = c_n f_n(x) + d_n f_{n-1}(x) $$ 

Then we can use intermediate clenshaw coefficients $b_n$ and then use them to compute $S(x)$.

Here $b_n$ follows the following recursion function,

$$ b_n = a_n + c_n b_{n+1} + d_{n+1} b_{n+2} $$

Clenshaw algorithm is numerically stable since it uses backward recursion

```python
def clenshaw(N, a, c, d, f0, f1):
    b = np.zeros(N+3)
    for i in range(N, -1, -1):
        b[i] = a[i] + c[i]*b[i+1] + d[i]*b[i+2]
        
    return f0*a[0] + f1*b[1] + d[1]*f0*b[2]
    
```

```python
def chebyshev_coef(n, f, large_n=10000):
    k = np.array(range(large_n))
    c = (2/large_n)*np.array([np.sum(f(np.cos(np.pi*(k+0.5)/large_n))*np.cos(np.pi*j*(k+0.5)/large_n)) for j in range(large_n)])
    c[0] /= 2
    return c[:n]

def chebyshev_func(a, x):
    n = len(a)
    y1 = np.array([np.dot(a, np.array([chebyt(order)(i) for order in range(n)])) for i in x])
    return y1

y = lambda x: np.exp(x)
```

### Direct Summation vs Clenshaw

```python
def direct_method(a, x, t0, t1):
    N = len(a)
    y = np.zeros(len(x))
    prev, prev_prev = t1(x), t0(x)
    for i in range(N):
        if i==0:
            y += a[0]*t0(x)
            continue
        elif i==1:
            y += a[1]*t1(x)
            continue
        else:
            f = 2*x*prev - prev_prev
            y += a[i]*f
            prev, prev_prev = f, prev
            
    return y
```

```python
N = 20

x = np.linspace(-1, 1, 1000)
y = lambda x: np.exp(x)
a = chebyshev_coef(N, y)
```

```python
y2 = direct_method(a, x, chebyt(0), chebyt(1))
y3 = np.array([clenshaw(N-1, a, 2*i*np.ones(N), -1*np.ones(N), 1, i) for i in x])
# y4 = chebyshev_func(a, x)
```

```python
plt.plot(x, y(x), label='True Function - $\exp(x)$')
plt.plot(x, y2, label='Direct Method')
plt.plot(x, y3, label='Clenshaw')
# plt.plot(x, y4, label='Exact Chebyshev')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

### Error Analysis

```python
N = 20

x = np.linspace(-1, 1, 1000)
y = lambda x: np.exp(x)
a = chebyshev_coef(N, y)


y2 = direct_method(a, x, chebyt(0), chebyt(1))
y3 = np.array([clenshaw(19, a, 2*i*np.ones(20), -1*np.ones(20), 1, i) for i in x])

err_direct = abs(y2-y(x))
err_clenshaw = abs(y3-y(x))
```

```python
plt.semilogy(x, err_direct, label='Direct Method')
plt.semilogy(x, err_clenshaw, label='Clenshaw')
# plt.semilogy(x, err_exact)
plt.legend()
plt.grid()
plt.grid(which='minor', ls=':',  lw=0.5)
plt.title("Error vs x")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.minorticks_on()
plt.show()
```

### Mean and Standard Deviation of Errors for various orders

```python
n = list(range(2, 100))
mean_errs_ch = []
std_errs_ch = []
mean_errs_clenshaw = []
std_errs_clenshaw = []
aN = chebyshev_coef(100, y)

for N in n:
    a = aN[:N]

    y2 = direct_method(a, x, chebyt(0), chebyt(1))
    y3 = np.array([clenshaw(N-1, a, 2*i*np.ones(N), -1*np.ones(N), chebyt(0)(i), chebyt(1)(i)) for i in x])

    err_direct = abs(y2-y(x))
    err_clenshaw = abs(y3-y(x))
    
    mean_errs_ch.append(np.mean(err_direct))
    std_errs_ch.append(np.std(err_direct))
    mean_errs_clenshaw.append(np.mean(err_clenshaw))
    std_errs_clenshaw.append(np.std(err_clenshaw))
```

```python
plt.semilogy(n, mean_errs_ch)
plt.fill_between(n, np.array(mean_errs_ch)+0.5*np.array(std_errs_ch), np.array(mean_errs_ch)-0.5*np.array(std_errs_ch), alpha=0.3)
plt.semilogy(n, mean_errs_clenshaw)
plt.fill_between(n, np.array(mean_errs_clenshaw)+0.5*np.array(std_errs_clenshaw), np.array(mean_errs_clenshaw)-0.5*np.array(std_errs_clenshaw), alpha=0.3)
plt.grid()
plt.grid(which='minor', ls=':', lw=0.5)
plt.minorticks_on()
plt.title("Mean error vs N (no. of terms)")
plt.xlabel("N")
plt.ylabel("Mean Error")
plt.show()
```

### Using Fourier Series

```python
# Coefficients computed analytically

def An(n):
    a = lambda n: (np.e**2 - 1)*pow(-1, n)/(np.e*(np.pi**2*n**2+1))
    res = a(np.array(range(0, n)))
    return res

def Bn(n):
    a = lambda n: -(np.e**2 - 1)*pow(-1, n)/(np.e*(np.pi**2*n**2+1))*np.pi*n
    res = a(np.array(range(1, n)))
    return res
```

```python
x = np.linspace(-1, 1, 1000)
y = lambda x: np.exp(x)
```

```python
def direct_method(n, x):
    an = An(n)
    bn = Bn(n)
    p0 = 1
    p1_cos = np.cos(x*np.pi)
    p1_sin = np.sin(x*np.pi)
    prev_prev_c, prev_prev_s = p0, 0
    prev_c, prev_s = p1_cos, p1_sin
    res = an[0]/2 + an[1]*p1_cos + bn[0]*p1_sin
    
    for i in range(2,n):
        f_c = 2*np.cos(np.pi*x)*prev_c - prev_prev_c
        f_s = 2*np.cos(np.pi*x)*prev_s - prev_prev_s
        res += an[i]*f_c + bn[i-1]*f_s
        prev_prev_c, prev_c = prev_c, f_c
        prev_prev_s, prev_s = prev_s, f_s
        
    return res
```

```python
an = An(20)
an[0] /= 2
bn = np.concatenate(([0], Bn(20)))

y_f_clenshaw_cos = np.array([clenshaw(19, an, 2*np.cos(i*np.pi)*np.ones(20), -1*np.ones(20), 1, np.cos(np.pi*i)) for i in x])
y_f_clenshaw_sin = np.array([clenshaw(19, bn, 2*np.cos(i*np.pi)*np.ones(20), -1*np.ones(20), 0, np.sin(np.pi*i)) for i in x])
```

```python
y_f_clenshaw = y_f_clenshaw_cos + y_f_clenshaw_sin
y_f_direct = direct_method(20, x)
```

```python
plt.plot(x, y_f_direct, label='Direct Method')
plt.plot(x, y_f_clenshaw, label='Chebyshev Method')
plt.plot(x, y(x), label='Exact Function')
plt.grid()
plt.legend()
plt.grid(which='minor', ls=':', lw=0.5)
plt.title("Fourier Fit - (N=20)")
plt.minorticks_on()
plt.show()
```

```python
err_direct = abs(y_f_direct-y(x))
err_clenshaw = abs(y_f_clenshaw-y(x))

plt.semilogy(x, err_direct, label='Direct Method')
plt.semilogy(x, err_clenshaw, label='Clenshaw')
# plt.semilogy(x, err_exact)
plt.legend()
plt.grid()
plt.grid(which='minor', ls=':',  lw=0.5)
plt.title("Error vs x - (N=20)")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.minorticks_on()
plt.show()
```

```python
n = list(range(2, 100))
mean_errs_fo = []
std_errs_fo = []
mean_errs_fourier = []
std_errs_fourier = []

for N in n:
    an = An(N)
    bn = Bn(N)
    an[0] /= 2
    bn = np.concatenate(([0], bn))

    y2 = direct_method(N, x)
    y_f_c = np.array([clenshaw(N-1, an, 2*np.cos(i*np.pi)*np.ones(N), -1*np.ones(N), 1, np.cos(np.pi*i)) for i in x])
    y_f_s = np.array([clenshaw(N-1, bn, 2*np.cos(i*np.pi)*np.ones(N), -1*np.ones(N), 0, np.sin(np.pi*i)) for i in x])
    y3 = y_f_c + y_f_s

    err_direct = abs(y2-y(x))
    err_clenshaw = abs(y3-y(x))
    
    mean_errs_fo.append(np.mean(err_direct))
    std_errs_fo.append(np.std(err_direct))
    mean_errs_fourier.append(np.mean(err_clenshaw))
    std_errs_fourier.append(np.std(err_clenshaw))
```

```python
plt.semilogy(n, mean_errs_fo, label='Direct')
plt.fill_between(n, np.array(mean_errs_fo)+0.5*np.array(std_errs_fo), np.array(mean_errs_fo)-0.5*np.array(std_errs_fo), alpha=0.3)
plt.semilogy(n, mean_errs_fourier, label='Clenshaw')
plt.fill_between(n, np.array(mean_errs_fourier)+0.5*np.array(std_errs_fourier), np.array(mean_errs_fourier)-0.5*np.array(std_errs_fourier), alpha=0.3)
plt.xlabel("Order")
plt.ylabel("Mean Error")
plt.title("Mean Error vs No. of Terms")
plt.legend()
plt.show()
```
