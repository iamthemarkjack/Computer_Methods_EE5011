```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn, yn
```

```python
x = 1
N = 26

# exact values
j_exact = np.array([jn(n,x) for n in range(0,N)])
y_exact = np.array([yn(n,x) for n in range(0,N)])

# jn in forward direction
j_for = [jn(0,x),jn(1,x)]

for n in range(2,N):
    j = (2*n/x)*j_for[-1] - j_for[-2]
    j_for.append(j)

j_for = np.array(j_for)

# yn in forward direction
y_for = [yn(0,x),yn(1,x)]

for n in range(2,N):
    y = (2*n/x)*y_for[-1] - y_for[-2]
    y_for.append(y)

y_for = np.array(y_for)

# jn in backward direction
j_bac = np.zeros(N)
j_bac[-1],j_bac[-2] = jn(N-1,x),jn(N-2,x)

for n in range(N-3,-1):
    j_bac[n] = (2*n/x)*j_bac[n+1] - j_bac[n+2]

# yn in backward direction
y_bac = np.zeros(N)
y_bac[-1],y_bac[-2] = yn(N-1,x),yn(N-2,x)

for n in range(N-3,-1):
    y_bac[n] = (2*n/x)*y_bac[n+1] - y_bac[n+2]
```

```python
errj_for = np.abs(j_for-j_exact)
errj_bac = np.abs(j_bac-j_exact)
erry_for = np.abs(y_for-y_exact)
erry_bac = np.abs(y_bac-y_exact)
n_vals = range(0,N)
plt.semilogy(n_vals,errj_for,label='Jn_forward')
plt.semilogy(n_vals,errj_bac,label='Jn_backward')
plt.semilogy(n_vals,erry_for,label='Yn_forward')
plt.semilogy(n_vals,erry_bac,label='Yn_backward')
plt.legend()
plt.show()
```
