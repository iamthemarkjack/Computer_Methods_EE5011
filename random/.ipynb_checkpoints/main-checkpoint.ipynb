```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2, norm, uniform, ecdf, kstest
from mpl_toolkits.mplot3d import Axes3D
from scipy.special import kolmogorov, chdtrc
```

```python
def gen_norm(n,seed=0):
    np.random.seed(seed)
    x1,x2 = np.random.rand(n,2).T
    y1 = np.sqrt(-2*np.log(x1))*np.cos(2*np.pi*x2)
    return y1
```

```python
# plotting the histogram for mu=0,sigma=1

for N in np.logspace(1,5,5,dtype=int):
    xx = np.linspace(-4,4,1000)
    yy = 1/(np.sqrt(2*np.pi))*np.exp(-xx**2/2)
    rand_numbers = gen_norm(N,26)
    plt.plot(xx,yy,c='black',lw=2)
    plt.hist(rand_numbers,bins='auto',density=True,color='red')
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.title(f'Density vs (x) for N = {N}')
    plt.show()
```

# Chi-Squared Test

```python
vals = []
bin_size = 0.1
bins = np.arange(-4,4,bin_size)
n_bins = len(bins)-1
for i in range(1000):
    rand = gen_norm(10000,i)
    observed,_ = np.histogram(rand,bins=bins)
    expected = np.diff(norm.cdf(bins))*len(rand)
    chi_squared = np.sum((observed-expected)**2/expected)
    vals.append(chi_squared)
x_upper = 200
xx = np.linspace(0,x_upper,1000)
df = n_bins-1
yy = chi2.pdf(xx,df,0.01)
plt.hist(vals,bins=100,density=True,color='blue')
plt.plot(xx,yy,color='black',lw=3)
plt.xlim([0,x_upper])
plt.show()
```

```python
vals = []
n_bins = 80
for i in range(10000):
    rand = gen_norm(1000,i)
    bins = norm.ppf(np.linspace(0,1,n_bins+1))
    observed,_ = np.histogram(rand,bins=bins)
    expected = np.diff(norm.cdf(bins))*len(rand)
    chi_squared = np.sum((observed-expected)**2/expected)
    vals.append(chi_squared)
xx = np.linspace(0,150,1000)
df = n_bins-1
yy = chi2.pdf(xx,df)
plt.hist(vals,bins=100,density=True,color='blue')
plt.plot(xx,yy,color='black',lw=3)
plt.xlim([0,200])
plt.show()
```

```python
def chi_square_norm_test(data,n_bins,alpha):
    bins = np.linspace(-4,4,n_bins+1)
    # bins = norm.ppf(np.linspace(0,1,n_bins+1))
    observed,_ = np.histogram(data,bins=bins)
    expected = np.diff(norm.cdf(bins))*len(data)
    chi_squared = np.sum((observed-expected)**2/expected)
    df = n_bins-1
    critical_val = chi2.ppf(1-alpha,df)
    p_val = chdtrc(df,chi_squared)
    return observed,expected,chi_squared,critical_val,p_val
```

```python
n_deviates = 100000
n_bins = 5
alpha = 0.05
```

### Normal Test

```python
rand = gen_norm(n_deviates)
observed,expected,chi_squared,critical_val,p_val = chi_square_norm_test(rand,n_bins,alpha)
```

```python
print('The value of Chi-Square Test Statistic : ',chi_squared)
print('p value : ',p_val)
print(f'The critical value for significance level of {100*alpha}% : ',critical_val)
```

```python
width = 0.3
ind = np.arange(len(expected))
plt.figure(figsize=(10,6))
plt.bar(ind,expected,width=width,label='Expected')
plt.bar(ind+width,observed,width=width,label='Observed')
plt.xticks(ind + width / 2, [f'bin_{i}' for i in range(n_bins)])
plt.legend(loc='best')
plt.xticks(rotation = -90)
plt.show()
```

```python
n_trys = 100
res = []
for seed in range(n_trys):
    rand = gen_norm(n_deviates,seed)
    observed,expected,chi_squared,critical_val,p_val = chi_square_norm_test(rand,n_bins,alpha)
    if p_val > alpha:
        res.append(1)
    else:
        res.append(0)
fail_to_reject_count = sum(res)
reject_count = len(res) - fail_to_reject_count
labels = ['Fail to Reject', 'Reject']
counts = [fail_to_reject_count, reject_count]
plt.bar(labels, counts, color=['blue', 'red'], alpha=0.7)
for i, count in enumerate(counts):
    plt.text(i, count + 0.5, str(count), ha='center', fontsize=12)
plt.title('Chi-Square Test Results')
plt.ylabel('Number of Trials')
plt.xlabel('Test Decision')
plt.show()
```

# Monte Carlo Integration

```python
def u(x,y):
    alpha = np.pi*np.sin(10*(np.sqrt(x**2 + y**2) - 0.5))
    u = x*np.cos(alpha) + y*np.sin(alpha) - 0.5
    return u

def v(x,y):
    alpha = np.pi*np.sin(10*(np.sqrt(x**2 + y**2) - 0.5))
    v = -x*np.sin(alpha) + y*np.cos(alpha) - 0.5
    return v

def f(x,y):
    return u(x,y)**2 + v(x,y)**2
```

```python
x = np.linspace(-4, 4, 1000)
y = np.linspace(-4, 4, 1000)
X, Y = np.meshgrid(x, y)

Z = (np.abs(f(X, Y)) <=1).astype(int)

plt.figure(figsize=(7,5))
cp = plt.contourf(X, Y, Z, levels=10, cmap='plasma')
plt.colorbar(cp)
plt.title("Contour plot of f(x, y)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', antialiased=False)
ax.set_title("3D Surface Plot of f(x, y)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.set_zlim(0,10)
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
plt.show()
```

```python
def monte_carlo_int(n):
    countin = 0
    countout = 0
    for i in range(n):
        a,b=np.random.rand(2)*4-2
        c=f(a,b)
        if abs(c)<1:
            countin+=1
        else:
            countout+=1
    return 16*countin/(countin+countout)
```

```python
vals = []
n_points = np.arange(10,10**4,100)
for i in range(10):
    vals.append(np.array([monte_carlo_int(n) for n in n_points]))
vals = np.array(vals)
```

```python
plt.figure(figsize=(10,6))
plt.semilogx(n_points,np.mean(vals,axis=0))
plt.fill_between(n_points,np.mean(vals,axis=0)-np.std(vals,axis=0),np.mean(vals,axis=0)+np.std(vals,axis=0),alpha=0.4)
plt.xscale('log')
plt.grid('both')
plt.xlabel('n_points')
plt.ylabel('Integral_val')
plt.show()
```

```python
final_val = np.mean(vals[:,-1])
final_std = np.std(vals[:,-1])

print('The Final Integral Value is : ',final_val)
print('Standard Deviation : ',final_std)
```
