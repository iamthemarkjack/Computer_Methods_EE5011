## Assignment - 2 : Spline Interpolation

```python
import numpy as np
import scipy as sp
from scipy.special import jv
from scipy.interpolate import CubicSpline

import matplotlib.pyplot as plt
```

```python
# Helper Functions

def func(x):
    return pow(x, 1+jv(0, x))/np.sqrt((1+100*x*x)*(1-x))


def deriv_1d(x):
    return (-jv(1, x)*np.log(x) + (1+jv(0, x))/x + (300*x*x - 200*x +1)/(2*(1+100*x*x)*(1-x)))*func(x)
```

```python
deriv_1d(0.1), deriv_1d(0.9)
```

#### Question 1 & 2

Analyzing the behaviour of the function from **0.1** to **0.9**

```python
x = np.arange(0.1, 0.95, 0.05)
y = func(x)


xx = np.linspace(0.1, 0.9, endpoint=True)
yy = func(xx)

plt.plot(x, y, 'ro', label='Data Points')
plt.plot(xx, yy, label='True Function')
plt.xlabel('x')
plt.ylabel('y')
plt.title('True Function Behaviour')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```

The given function is analytic in the given region

The function has **3** poles, at $z=\pm 0.01i$ and at $z=1$.
The domain of the function : $x \in (0, 1)$

Radius of convergence at $x=0.1$ and at $x=0.9$ are both **0.1**

#### Question 3

Using Cubic Spline interpolation to interpolate **n** equally spaced points and trying to estimate the optimal *n* for achieving 6-digit accuracy.

We will take the second derivatives at the boundary to be $$y'' = 0$$

```python
# Helper Functions

def cubic_spline_wrapper(x, y, n, xx, method=None, fd=0, ed=0):
    if method:
        cs = CubicSpline(x, y, bc_type=method)
    else:
        cs = CubicSpline(x, y, bc_type=((1, fd), (1, ed)))
        
    yy = cs(xx)

    return yy
```

```python
max_z = []

for n in range(5, 1000, 50):
    x = np.linspace(0.1, 0.9, n, endpoint=True)
    y = func(x)

    xx = np.linspace(0.1, 0.9, n*10, endpoint=True)
    yy_true = func(xx)
    
    yy = cubic_spline_wrapper(x, y, n, xx, "natural")
    z = abs(yy_true-yy)

    if n == 105:
        plt.semilogy(xx, z, label='Error')
        plt.grid()
        plt.grid(which='minor', linestyle=":", linewidth=0.5)
        plt.minorticks_on()
        plt.title("Error Plot for n = 105 points")
        plt.show()

    max_z += [np.max(z)]
    
plt.semilogy(range(5, 1000, 50), max_z, marker='o', color='black', markerfacecolor='red')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Error vs n - $(y''=0)$")
plt.show()
```

#### Question 7

Instead of choosing equal spacing, could we do something better ?

> Proposal :
>
> We could first sample uniformly spaced points and multiply them with the weights of the values obtained to better capture the function

```python
def chebyshev_nodes(n, a, b):
    theta = np.linspace(0, np.pi, n, endpoint=True)
    x = np.cos(theta)[::-1]
    x = (b-a)*0.5*x + (b+a)*0.5
    # print(x)

    return x


def derive_nodes(n, a, b, deriv):
    alpha = (b-a)/n*0.99
    x = np.linspace(a, b, n, endpoint=True)
    x_mod = np.zeros(n)
    x_mod[0] = a
    x_mod[-1] = b
    for idx, i in enumerate(x[1:-1]):
        der = deriv(i)
        if abs(der) <= alpha:
            i += der

        else:
            if der*alpha > 0:
                i += alpha
            else:
                i -= alpha

        x_mod[idx+1] = i

    print(x_mod)
    return x_mod
```

```python
n = 22

# fd = 100*deriv_1d(0.1)
# ed = 100*deriv_1d(0.9)


# x = np.linspace(0.1, 0.9, n, endpoint=True)
x = chebyshev_nodes(n, 0.1, 0.9)
# x = derive_nodes(n, 0.1, 0.9, deriv_1d)
y = func(x)

xx = np.linspace(0.1, 0.9, 1000, endpoint=True)
yy_true = func(xx)

yy = cubic_spline_wrapper(x, y, n, xx, "not-a-knot")
z = abs(yy_true-yy)


plt.plot(x, y, 'ro', label='Data Points')
plt.plot(xx, yy, 'black', label='Interpolated Function')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.title("Using Chebyshev Points")
plt.show()

plt.semilogy(xx, z, label='Error')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.title("Error with Chebyshev Points")
plt.show()

############################################################

x = np.linspace(0.1, 0.9, n, endpoint=True)
y = func(x)

xx = np.linspace(0.1, 0.9, 1000, endpoint=True)
yy_true = func(xx)

yy = cubic_spline_wrapper(x, y, n, xx, "not-a-knot")
z = abs(yy_true-yy)


plt.plot(x, y, 'ro', label='Data Points')
plt.plot(xx, yy, 'black', label='Interpolated Function')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.title("Using Equally Spaced Points")
plt.show()

plt.semilogy(xx, z, label='Error')
plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.title("Error with Equi-spaced Points")
plt.show()
```

### Sympy Derivatives

```python
from sympy import *
from sympy.functions.special.bessel import besselj
```

```python
x = symbols('x')
expr_diff = pow(x, 1+besselj(0, x))/((1+100*x*x)*(1-x))**(0.5)
```

```python
expr_diff
```

```python
exprs = []
x_ = np.linspace(0.1, 0.9, 100)
y = np.zeros((4, x_.size))

for order in range(0, 4):
    expr_diff = Derivative(expr_diff, x).doit()
    # print(f'{order+1} derivative - {expr_diff}')
    exprs.append(expr_diff)
    for idx, el in enumerate(x_):
        y[order, idx] = expr_diff.subs(x, el)
```

```python
for order in range(1, 5):
    plt.plot(x_, y[order-1], label='$y$' + "'"*order)
    # plt.semilogy(x_, y[order-1], label='$y$' + "'"*order)

plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.title("Derivatives of $y(x)$")
plt.xlabel("x")
plt.legend()
plt.show()

for order in range(1, 5):
    # plt.plot(x_, y[order-1], label='$y$' + "'"*order)
    plt.semilogy(x_, y[order-1], label='$y$' + "'"*order)

plt.grid()
plt.grid(which='minor', linestyle=':', linewidth=0.5)
plt.minorticks_on()
plt.title("Derivatives of $y(x)$")
plt.xlabel("x")
plt.legend()
plt.show()


```

```python
def approx_func(x, order):
    return 0.02/(1-x)**(order*3/2)
```

```python
y1 = approx_func(x_, 1)
y2 = approx_func(x_, 2)
y3 = approx_func(x_, 3)
y4 = approx_func(x_, 4)
```

```python
plt.plot(x_, y1, label="$y'$")
plt.plot(x_, y2, label="$y''$")
plt.plot(x_, y3, label="$y'''$")
plt.plot(x_, y4, label="$y''''$")

plt.grid()
plt.grid(which='minor', linestyle=":", linewidth=0.5)
plt.minorticks_on()
plt.xlabel("x")
plt.legend()
plt.title("Approximation of the derivative functions")
plt.show()
```

```python
plt.semilogy(x_, y1, label="$y'$")
plt.semilogy(x_, y2, label="$y''$")
plt.semilogy(x_, y3, label="$y'''$")
plt.semilogy(x_, y4, label="$y''''$")

plt.grid()
plt.grid(which='minor', linestyle=":", linewidth=0.5)
plt.minorticks_on()
plt.xlabel("x")
plt.legend()
plt.title("Approximation of the derivative functions")
plt.show()
```

### Error vs Spacing

```python
h = np.logspace(-3, -0.5, 10, endpoint=False)
```

```python
xx = np.linspace(0.1, 0.9, 1000, endpoint=True)
yy_true = np.sin(xx)
max_z = []

for spacing in h:
    n = int(0.8/spacing)
    x = np.linspace(0.1, 0.9, n, endpoint=True)
    y = np.sin(x)

    # yy = cubic_spline_wrapper(x, y, n, xx, "nautral")
    yy = CubicSpline(x, y, bc_type="natural")(xx)
    z = abs(yy - yy_true)
    max_z += [z.max()]

plt.loglog(h, max_z)
plt.grid()
plt.title("Error vs spacing")
plt.xlabel("Spacing")
plt.ylabel("Error")
plt.show()
```
